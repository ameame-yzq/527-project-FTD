```python
import pandas as pd

# read csv file
data = pd.read_csv('archive_export (1).csv')
```

```python
# show data head
data.head()
```

```python
# shape info
shape_info = data.shape
shape_info
```

```python
# get column name
columns_info = data.columns
columns_info
```

```python
# get every column info 
info_dict = {}
for column in data.columns:
    non_null_count = data[column].count()
    column_dtype = data[column].dtype
    info_dict[column] = {"non_null_count": non_null_count, "dtype": column_dtype}
```

```python
# get stasitics info
describe_info = data.describe()
describe_info
```

```python
# Create a DataFrame to display the results

# Convert column information to DataFrame
columns_df = pd.DataFrame(info_dict).transpose()

# Create a DataFrame for descriptive statistical information
describe_info_df = describe_info.transpose()[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]

# Merge column information and descriptive statistical information
combined_df = pd.concat([columns_df, describe_info_df], axis=1).fillna('-')

# Add the data shape and column names to the top of the DataFrame
shape_df = pd.DataFrame({
    'non_null_count': [shape_info[0], shape_info[1]],
    'dtype': ['-', '-'],
}, index=['Number of Rows', 'Number of Columns'])

final_df = pd.concat([shape_df, combined_df])

final_df
```

```python
# Convert the column names to a DataFrame where each column name is in a separate row
column_names_df = pd.DataFrame(columns_info, columns=['Column Names'])
column_names_df
```
